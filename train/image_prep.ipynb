```python
import csv
import os
import shutil, sys
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
```

```python
folder = [name for name in os.listdir("./images")]
print(len(folder))
```

```python
filename = "fastfood.csv"
 
# initializing the titles and rows list
fields = []
rows = []

csvdata = []
 
# reading csv file
with open(filename, 'r') as csvfile:
    # creating a csv reader object
    csvreader = csv.reader(csvfile)
 
    # extracting field names through first row
    fields = next(csvreader)
 
    # extracting each data row one by one
    for row in csvreader:
        rows.append(row[0] + ' ' + row[1])
        csvdata.append([row[0] + ' ' + row[1]] + row[2:])
        

classes = set()
for row in rows:
    if row == 'Burger King Crispy Chicken Jr':
        print('yuh')
    if row.find('/') < 0 and row.find('Sonic') < 0 and row.find('Dairy Queen') < 0 and row.find('"') < 0:
        if row[-1] == '.':
            classes.add(row.replace('Â®', '').replace('â€', '')[:-1])
        else:
            classes.add(row.replace('Â®', '').replace('â€', ''))
classes = list(classes)
```

```python
fields = []
csvdata = []
 
# reading csv file
with open(filename, 'r') as csvfile:
    # creating a csv reader object
    csvreader = csv.reader(csvfile)
 
    # extracting field names through first row
    fields = next(csvreader)
 
    # extracting each data row one by one
    for row in csvreader:
        clean = row[0] + ' ' + row[1]
        if clean.find('/') < 0 and clean.find('Sonic') < 0 and clean.find('Dairy Queen') < 0 and clean.find('"') < 0:
            if clean[-1] == '.':
                clean = clean.replace('Â®', '').replace('â€', '')[:-1]
            else:
                clean = clean.replace('Â®', '').replace('â€', '')
            csvdata.append([clean] + row[2:])
csvdata.sort()
```

```python
os.chdir('images')
print(os.getcwd())
if os.path.isdir('train') is False: 
    os.mkdir('train')
    os.mkdir('valid')
    os.mkdir('test')

    for label in classes:
        print('Current Label: ' + label)
        batches = ImageDataGenerator(rotation_range=10, width_shift_range=0.1, height_shift_range=0.1, shear_range=0.15, zoom_range=0.1, 
        channel_shift_range=10.0, horizontal_flip=True).flow_from_directory(
            '',
            batch_size=40,
            save_to_dir=label,
            save_prefix='aug-',
            classes=classes,
            interpolation='nearest',
            shuffle=False,
            keep_aspect_ratio=False
        )
        next(batches)

        shutil.move(f'{label}', 'train')
        os.mkdir(f'valid/{label}')
        os.mkdir(f'test/{label}')
    

        valid_samples = random.sample(os.listdir(f'train/{label}'), 10)
        for j in valid_samples:
            shutil.move(f'train/{label}/{j}', f'valid/{label}')

        test_samples = random.sample(os.listdir(f'train/{label}'), 10)
        for k in test_samples:
            shutil.move(f'train/{label}/{k}', f'test/{label}')
        print('DONE')
os.chdir('..')
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 10, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```
